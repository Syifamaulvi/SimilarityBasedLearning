```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Data
data = pd.DataFrame({
    'Speed': [2.50,3.75,2.25,3.25,2.75,4.50,3.50,3.00,4.00,4.25,
              2.00,5.00,8.25,5.75,4.75,5.50,5.25,7.00,7.50,7.25],

    'Agility': [6.00,8.00,5.50,8.25,7.50,5.00,5.25,3.25,4.00,3.75,
                2.00,2.50,8.50,8.75,6.25,6.75,9.50,4.25,8.00,5.75],

    'Draft': ["no","no","no","no","no","no","no","no","no","no",
              "no","no","no","yes","yes","yes","yes","yes","yes","yes"]
})

print(data)
```

```python
#Scatterplot
yes = data[data['Draft'] == 'yes']
no = data[data['Draft'] == 'no']

plt.figure()

plt.scatter(yes['Speed'], yes['Agility'], marker='x', label='Draft YES')
plt.scatter(no['Speed'], no['Agility'], marker='o', label='Draft NO')


plt.xlabel("Speed")
plt.ylabel("Agility")
plt.title("Scatter Plot: Speed vs Agility")
plt.legend()

plt.show()
```

```python
#Measuring Similarity Using Distance Matrix

X = data[['Speed', 'Agility']].values
n = len(X)
```

```python
#Eculidien Matrix
distance_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distance_matrix[i, j] = np.sqrt(np.sum((X[i] - X[j])**2))

distance_df = pd.DataFrame(distance_matrix)

print(distance_df.round(2))
```

```python
distance_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distance_matrix[i, j] = np.sum(np.abs(X[i] - X[j]))

distance_df = pd.DataFrame(distance_matrix)
print(distance_df.round(2))
```

```python
# MINKOWSKI MATRIX
p = 3
distance_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distance_matrix[i, j] = np.sum(np.abs(X[i] - X[j])**p)**(1/p)


distance_df = pd.DataFrame(distance_matrix)

print(distance_df.round(2))
```

```python
# QUERY INSTANCE (DATA BARU)

query = pd.DataFrame({
    'Speed': [6.75],
    'Agility': [3.00]
})

print(query)
```

```python
# HITUNG JARAK EUCLIDEAN

data['Distance'] = np.sqrt(
    (data['Speed'] - query.loc[0, 'Speed'])**2 +
    (data['Agility'] - query.loc[0, 'Agility'])**2
)

print(data[['Speed', 'Agility', 'Draft', 'Distance']].round(4))
```

```python
# URUTKAN BERDASARKAN JARAK

nearest = data.sort_values(by='Distance')

print(nearest[['Speed', 'Agility', 'Draft', 'Distance']].head().round(4))
```

```python
# AMBIL 1 TETANGGA TERDEKAT

nearest_1 = nearest.iloc[0]
prediction = nearest_1['Draft']

print("Nearest Neighbor:")
print(nearest_1[['Speed', 'Agility', 'Draft', 'Distance']])
print("Prediksi Draft untuk query instance =", prediction)
```

```python
data_sorted = data.sort_values(by='Distance')
print(data_sorted)
```

```python
k = 3
neighbors = data_sorted.head(k)
prediction = neighbors['Draft'].mode()[0]

print(neighbors[['Speed','Agility','Draft','Distance']])
print("Prediksi =", prediction)
```
